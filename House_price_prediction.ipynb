```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['street']=le.fit_transform(df['street'])
le1=LabelEncoder()
df['city']=le1.fit_transform(df['city'])
```

```python
features=['statezip','country','date']
for i in features:
  df.drop([i],axis=1,inplace=True)
```

```python
import seaborn as sns
sns.heatmap(df.corr())
```

```python
x=df.iloc[:,1:]
x
```

```python
y=df.iloc[:,:1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
y_pred_mlr=mlr.predict(x_test)
y_pred_mlr
```

```python
from sklearn.metrics import r2_score
print("r2score:",r2_score(y_test,y_pred_mlr))
```

```python
import matplotlib.pyplot as plt
fig,axs = plt.subplots(1,2, figsize=(10, 5))
axs[0].hist(y_test,bins=20,edgecolor='black')
axs[0].set_xlabel('ACTUAL PRICE')
axs[0].set_ylabel('FREQUENCY')
axs[0].set_title('ACTUAL PRICE DISTRIBUTION')
axs[1].hist(y_pred_mlr,bins=20,edgecolor='black')
axs[1].set_xlabel('PREDICTED PRICE')
axs[1].set_ylabel('FREQUENCY')
axs[1].set_title('PREDICTED PRICE DISTRIBUTION')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
y_pred_rfr=rfr.predict(x_test)
print("r2score:",r2_score(y_test,y_pred_rfr))
```

```python
fig,axs = plt.subplots(1,2, figsize=(10, 5))
axs[0].hist(y_test,bins=20,edgecolor='black')
axs[0].set_xlabel('ACTUAL PRICE')
axs[0].set_ylabel('FREQUENCY')
axs[0].set_title('ACTUAL PRICE DISTRIBUTION')
axs[1].hist(y_pred_rfr,bins=20,edgecolor='black')
axs[1].set_xlabel('PREDICTED PRICE')
axs[1].set_ylabel('FREQUENCY')
axs[1].set_title('PREDICTED PRICE DISTRIBUTION')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
y_pred_dtr=dtr.predict(x_test)
print("r2score:",r2_score(y_test,y_pred_dtr))
```

```python
fig,axs = plt.subplots(1,2, figsize=(10, 5))
axs[0].hist(y_test,bins=20,edgecolor='black')
axs[0].set_xlabel('ACTUAL PRICE')
axs[0].set_ylabel('FREQUENCY')
axs[0].set_title('ACTUAL PRICE DISTRIBUTION')
axs[1].hist(y_pred_dtr,bins=20,edgecolor='black')
axs[1].set_xlabel('PREDICTED PRICE')
axs[1].set_ylabel('FREQUENCY')
axs[1].set_title('PREDICTED PRICE DISTRIBUTION')
plt.show()
```
